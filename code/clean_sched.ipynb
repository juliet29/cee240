```python
def sort_activities_in_area(rows):
    increm = df.iloc[rows[0]:rows[1], 0:2]
    act = list(increm["activities"])
    ixes = list(pd.to_numeric(increm["index"]))
    sort_ix = np.argsort(ixes)
    increm_1_ordered = [act[i] for i in sort_ix]
    df["act_ordered"][rows[0]:rows[1]] = increm_1_ordered
    return df

    
```

```python
row_nums = [[3,59], [62,87], [89,113], [115,132], [134,150], [152,169], [172, 186], [188,193]]
```

```python
# create column for ordered activities 
df["act_ordered"] = ""
for i in row_nums:
    print(i)
    i[0] = i[0] -3
    i[1] = i[1] -2
    sort_activities_in_area(i)
```
